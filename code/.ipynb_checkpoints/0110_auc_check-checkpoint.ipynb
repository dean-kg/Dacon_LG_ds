## auc 테스트 코드

```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

mu_f, sigma_f = 0.5, 0.1 
mu_t, sigma_t = 0.5, 0.1


threshhold = 0.5

false_prob = np.random.normal(mu_f, sigma_f, 1000)
true_prob  = np.random.normal(mu_t,sigma_t, 1000) 
false_prob_filter =false_prob[np.where(false_prob>=0)]
true_prob_filter =true_prob[np.where(true_prob<=1)]


trues = np.append(np.zeros(len(false_prob_filter)),np.ones(len(true_prob_filter)))
preds = np.append(false_prob_filter,true_prob_filter)



count_f, bins_f, ignored_f = plt.hist(false_prob_filter, 30, density=True,label='false probs')
count_t, bins_t, ignored_t = plt.hist(true_prob_filter, 30, density=True,color='orange',label='ture probs')


plt.plot(bins_f, 1/(sigma_f * np.sqrt(2 * np.pi)) * 
np.exp( - (bins_f - mu_f)**2 / (2 * sigma_f**2) ), linewidth=2, color='r')

plt.plot(bins_t, 1/(sigma_t * np.sqrt(2 * np.pi)) * 
np.exp( - (bins_t - mu_t)**2 / (2 * sigma_t**2) ), linewidth=2, color='r')

plt.title("probs - models")
plt.xlabel('Probs')
plt.ylabel("Density")
plt.xlim(0,1)
plt.legend()
plt.show()

fpr, tpr, thresholds = metrics.roc_curve(trues, preds, pos_label=1)
roc_auc=metrics.auc(fpr, tpr)

import matplotlib.pyplot as plt
plt.title('Roc_curve')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

## 개념 정리

recall = 재현율 = 민감도 = sensitivity = True positive rate (양성      / 전체 양성)    
precision = 정밀도 =                 (양성      / 모델 양성 )    
specificity = 특이도                 (판정중 음성 / 전체음성)

민감도 <-> 특이도 상대 개념 (더해서 100% 아님)

통상적인 분류 모델의 경우 50%를 기준으로 판단한다. (threshhold)    
roc curve 의 x축은 1-specificity , y축은 recall(sensitivity) 값이다.    

학습시 데이터나 모델에 따라 threshhold 값이 0.5가 아닌 다른 값으로 지정하는 경우 존재
